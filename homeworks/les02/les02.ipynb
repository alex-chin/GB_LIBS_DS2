# Урок 2. Анализ данных и проверка статистических гипотез. Практическое задание

1. В чём различие между зависимыми и независимыми выборками?
2. Когда применяются параметрические статистические критерии, а когда — их непараметрические аналоги?

### 1. Зависимые и независимые выборки

В общем случае выборки являются зависимыми, если можно установить каждому случаю из одной выборки один и только один случай из второй выборки. Если такой связи не установлено, то выборки - независимые. 

### 2. Применение параметрических и непараметрических статистических критериев

Параметрические статистические критерии включают в расчет параметры вероятностного распределения признака - среднее и дисперия. Как правило это выподняется для вещественных признаков. Поэтому если параметрические статистические критерии не возможно применить по этим или другим причинам (не нормальное распределение, выбросы, менее 30 наблюдений), то применяют непараметрические статистические критерии. Они основаны на оперировании частотами или рангами.
