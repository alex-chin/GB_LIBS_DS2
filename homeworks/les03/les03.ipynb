### 1. Варианты усреднения для метрик качества классификации: micro, macro, weighted

**micro** 

При расчете **micro** не важен класс (фактически это accuracy), важен факт правильности классификации. 

>Используется если отсутствует дисбаланс в классах 

**macro**

При расчете **macro** считается **recall** для всех классов и усредняется и считается **precision** для всех классов и усредняется 

>Используется когда присутсвует дисбаланс в классах

 **weighted**

При расчете **weighted** каждая мера берется с нормировкой по количеству объектов в выборке. 

> Используется в случае если нам важна метрика для оценки класса с бо^льшим количеством объектов

В общем и целом в большинстве случаев надо ориентироваться на macro f1 т.к. эта метрика как правило самая маленькая.

### 2. Модели xgboost, lightgbm и catboost и их основные особенности

В общем и целом модели **xgboost, lightgbm и catboost** это вариации на построение эффективных решающих деревьев.

**xgboost** 

Особенность - вводит показатель *Похожесть* вместо Энтропии при оценке объектов в листе. Например для 1 объекта Похожесть = 1. Также по своему расчитывается прирост.

Кроме того введен ряд параметров ($\lambda$, $\gamma$, $\eta$) для тюнинга - в основном влияет на снижение переобучения.

**lightgbm** 

Для ускорения бустинга использует не все объекты на каждой иттерации - выбираются объекты с наибольшей ошибкой (GOSS). Также позволяет объединять разреженные данные - два признака в один путем добавления смещения (EFB). 

Построение дерева идет сначала в ширину - по листьям с максимальной ошибкой.

**catboost**

Алгоритм проводит обучение нескольких моделей за иттерацию и высчитывает остатки на других моделей, поэтому обучается долго. Для предотвращения переобучения на начальной стадии берется высокий уровень рандома (перемешивание). При этом все вопросы в дереве симметричные, поэтому обученный классификатор работает крайне эффективно. Кроме того в алгоритм заложены возможности эффективной работы с категориальными признаками.

>В общем и целом на начальном этапе возможно использование  lightgbm как наиболее быстрой, если наступает переобучение, то возможно использование xgboost c тюнингом. При особых требованиях на production и работе с категориями используем catboost
