# Урок 1. Введение в задачу классификации. Постановка задачи и подготовка данных. Практическое задание

1. Приведите по 2 примера, когда лучше максимизировать Precision, а когда Recall.
2. Почему мы используем F-меру, почему, например, нельзя просто взять среднее от Precision и Recall?

### Определения


<center>$\large precision = \LARGE \frac{TP}{TP \: + \: FP} \:\:\:\:\: \large recall = \LARGE \frac{TP}{TP \: + \: FN}$<center>

Мера точности (P, Precision) характеризует, сколько полученных от классификатора положительных ответов являются правильными. Чем больше точность, тем меньше число ложных попаданий.

Мера точности, однако, не дает представление о том, все ли правильные ответы вернул классификатор. 

Для этого существует так называемая мера полноты (R, recall).

Мера полноты характеризует способность классификатора «угадывать» как можно большее число положительных ответов из ожидаемых. Заметим, что ложно-положительные ответы никак не влияют на эту метрику.

Precision и Recall дают довольно исчерпывающую характеристику классификатора, причем «с разных углов». Обычно при построении подобного рода систем приходится все время балансировать между двумя этими метриками. 

Если пытаться повысить Recall, делая классификатор более «оптимистичным», это приводит к падению Precision из-за увеличения числа ложно-положительных ответов. 

Если же подкручивать классификатор, делая его более «пессимистичным», например, строже фильтруя результаты, то при росте Precision это вызовет одновременное падение Recall из-за отбраковки какого-то числа правильных ответов.

### 1. Примеры

**Use case 1.** Для определения сбоев в работе сетевой инфраструктуры используется детектор (классификатор). Стратегия его работы "оптимистическая" - все что не соответсвует нормальной работы, сети признается _аномалией_. 

>В этом случае выбираем модель с максимальной полнотой (max recall).

**Use case 2.** Для своевременной реакции на инциденты информационной безопасности необходимо провести строгую классификацию анномалий и выявления конкретного класса (классов) _инцидента_ для соотнесения с планом действий (playbook). Стратегия работы классификатора - пессимистичная для строгого отбора инцидентов.

> В этом случае выбираем модель с максимальной точностью (max precision).

**Use case 3.** Для работы системы общения на естественном языке необходимо определить момент обращения человека к системе. Классификатор на этом уровне реагирует на все шумы выявляя звуки похожие на голос. Стратегия его работы "оптимистическая" - весь разговор прослушивается - происходит выделение голоса от шума.

>В этом случае выбираем модель с максимальной полнотой (max recall).

**Use case 4.** При работе Виртуального помощника, модель в потоке диалога должна выявлять команды (указания) для совершения действий, вплоть до включения аппаратуры. Стратегия работы классификатора - пессимистичная для строгого отбора команд - Создай заметку, Позвони, Включи. 

>В этом случае выбираем модель с максимальной точностью (max precision).

### 2. Использование  F-меры

Precision и Recall характеристику классификатора с разных углов зрения. И бычно при построении модели приходится все время балансировать между двумя этими метриками.

Если пытаться повысить Recall, делая классификатор более «оптимистичным», это приводит к падению Precision из-за увеличения числа ложно-положительных ответов.

Если же подкручивать классификатор, делая его более «пессимистичным», например, строже фильтруя результаты, то при росте Precision это вызовет одновременное падение Recall из-за отбраковки какого-то числа правильных ответов.





Поэтому удобно для характеристики классификатора использовать одну величину, так называемую метрику F:


<center>$\LARGE F_{\beta} = (1+ \beta^2) \cdot \frac{precision \: \cdot \: recall}{(\beta^2 \cdot precision) \: + \: recall}$<center>

Коэффициент  β , позволяет балансировать вклад, вносимый точностью, тем самым акцентируя внимание на той метрике, максимизация которой для нас наиболее приоритетна.

При β=1 для нас будет важно влияние метрик в равной степени. Фактически это просто среднее гармоническое величин P и R. Метрика F1 достигает своего максимума 1 (100%), если P = R = 100%.

**Почему не подходит среднее**

Если Precision или Recall будут очень маленькими, то среднее не покажет этой плохой ситуации, а F-мера как раз-таки поможет в этом. (Пример: Precision = 0.7, Recall=0.1, среднее=0.4, F-score=0.175)
